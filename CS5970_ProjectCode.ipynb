```python
from inspect import modulesbyfile
import pandas as pd
# Here's the csv file containing the participant information
participant_data = pd.read_csv('drive/MyDrive/Colab Notebooks/Data/participants.tsv', delimiter='\t')
participant_ces = pd.read_csv('drive/MyDrive/Colab Notebooks/Data/CES-D.tsv', delimiter='\t')
# The following code obtains ndarrays of each characteristic and CES-D scores from each time point
# Get group in nparray
group = 'group'
group_data = participant_data[group].tolist()
group_data = np.array(group_data)
# Get age in nparray
age = 'age'
age_data = participant_data[age].tolist()
age_data = np.array(age_data)
# Get sex in nparray
sex = 'sex'
sex_data = participant_data[sex].tolist()
sex_data = np.array(sex_data)
# Get height in nparray
height = 'size'
height_data = participant_data[height].tolist()
height_data = np.array(height_data)
# Get weight in nparray
weight = 'weight'
weight_data = participant_data[weight].tolist()
weight_data = np.array(weight_data)
# Get CES-D scores from Time Point 1
ses1 = 'CES-D_1'
ses1_data = participant_ces[ses1].tolist()
ses1_data = np.array(ses1_data)
# Get CES-D scores from Time Point 2
ses2 = 'CES-D_2'
ses2_data = participant_ces[ses2].tolist()
ses2_data = np.array(ses2_data)
# Get CES-D scores from Time Point 3
ses3 = 'CES-D_3'
ses3_data = participant_ces[ses3].tolist()
ses3_data = np.array(ses3_data)
#print(ses3_data)
# Print all arrays
#print('group', group_data)
#print('ages', age_data)
#print('sex', sex_data)
#print('height', height_data)
#print('weight', weight_data)
#print(participant_data)

# Now let's group all the columns together in preparation for a matrix
participants = np.column_stack((group_data, age_data, sex_data, height_data, weight_data))
#print(participants)

# Now we need to separate the groups
# Get indices where 'group' is equal to 1
group_1_indices = np.where(participants[:, 0] == '1')

# Get indices where 'group' is equal to 2
group_2_indices = np.where(participants[:, 0] == '2')

# Create matrices for group 1 and group 2
group_1_matrix = participants[group_1_indices]
group_2_matrix = participants[group_2_indices]

# Compute the differences in CES-D scores after running intervention. Group 1 = t2-t1, Group 2 = t3-t2
group1ces_difference = ses2_data - ses1_data
group2ces_difference = ses3_data - ses2_data
# Now let's obtain the groups with their respective CES-D scores
group1ces_difference = np.column_stack((group1ces_difference, group_data))
group2ces_difference = np.column_stack((group2ces_difference, group_data))
group_1_indices_ces = np.where(group1ces_difference[:, 1] == 1)
group1ces_difference = group1ces_difference[group_1_indices_ces]
group_2_indices_ces = np.where(group2ces_difference[:, 1] == 2)
group2ces_difference = group2ces_difference[group_2_indices_ces]
group1ces_difference = np.delete(group1ces_difference, 1, axis=1)
group2ces_difference = np.delete(group2ces_difference, 1, axis=1)
# Now the two arrays contain the groups respective change in CES-D scores
# Used for pearson test
change_in_ces = np.concatenate((group1ces_difference,group2ces_difference))
# This contains the change in CES-D scores after running
change_in_ces=np.squeeze(change_in_ces)
#print(change_in_ces)
# now we obtain each matrix with their respective CES-D scores
group_1_matrix = np.column_stack((group_1_matrix, group1ces_difference))
group_2_matrix = np.column_stack((group_2_matrix, group2ces_difference))

#print('group1', group_1_matrix)
#print('group2', group_2_matrix)
# Now we have a matrix full of features and CES-D as the label
result_matrix = np.concatenate((group_1_matrix, group_2_matrix), axis=0)
result_matrix = np.delete(result_matrix, 0, axis=1)
#print(result_matrix)
# Change the M,F in the gender column to 1,0 respectively
gender_column = result_matrix[:, 1]
result_matrix[:, 1] = np.where(gender_column == 'M', 1, 0)
#print(result_matrix)
result_matrix = result_matrix.astype(float)
#print(result_matrix)
#print(np.shape(result_matrix))
n,p=np.shape(result_matrix)
#print(n,p)
# The resulting matrix contains the four characteristics of each participant as features: age,sex,height,weight.
# the label is the difference between after the running intervention and before it (after-before).
# Negative values means a decrease in CES-D scores (a decrease in depressive symptoms)
# Positive values means an increase in CES-D scores (an increase in depressive symptoms)

# import packages for machine learning
from sklearn.linear_model import LinearRegression
from sklearn.kernel_ridge import KernelRidge
from sklearn.linear_model import Ridge, Lasso, LassoCV, RidgeCV
from sklearn.metrics import make_scorer, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split, cross_val_predict, KFold, cross_val_score
import sklearn as sk

#result_matrix = np.delete(result_matrix,0,axis=1)
#result_matrix = np.delete(result_matrix,1,axis=1)

X=result_matrix[:,0:-1]
y=result_matrix[:,-1]

# Initialize the scaler
scaler = StandardScaler()
# Use K fold for better accuracy
skf = sk.model_selection.KFold(n_splits=5)
folds = skf.split(X,y)
train_idxs=[]
test_idxs=[]
for i, fold in enumerate(folds):
#  print("Processing fold %i" % (i + 1))
  # Here, we unpack fold (a tuple) to get the train and test indices
  train_idx, test_idx = fold
  train_idxs.append(train_idx)
  test_idxs.append(test_idx)
  #print("Train indices:", train_idx)
  #print("Test indices:", test_idx, end='\n\n')
clf=LinearRegression()
mse_cv=[]
# For loop for calculating each fold mse
for i in range(5):
  X_train = X[train_idxs[i]]
  Y_train = y[train_idxs[i]]
  X_test = X[test_idxs[i]]
  Y_test = y[test_idxs[i]]
  X_train_norm = scaler.fit_transform(X_train)
  X_test_norm = scaler.transform(X_test)
  clf.fit(X_train_norm, Y_train)
  preds = clf.predict(X_test_norm)
  mse_cv.append(mean_squared_error(Y_test, preds))
print('average mean after k fold with k=5:',np.mean(mse_cv))

model=LinearRegression()

# set training number and testing number
num_train = int(0.8*n)
num_test = int(0.2*n)

scaler.fit_transform(result_matrix)

# split data into training set and testing set
X_train = result_matrix[0:num_train,0:-1]
X_test = result_matrix[n-num_test:,0:-1]
Y_train = result_matrix[0:num_train,-1]
Y_test = result_matrix[n-num_test:,-1]

# Fit the model
model.fit(X_train, Y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the performance
mse = mean_squared_error(Y_test, y_pred)
r2 = r2_score(Y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
# Plot the predictions vs. actual values
plt.scatter(Y_test, y_pred)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('CES-D Scores')

# Fit a line of best fit using numpy's polyfit
slope, intercept = np.polyfit(Y_test, y_pred, 1)

# Plot the line of best fit
plt.plot(Y_test, slope * Y_test + intercept, color='red', label='Line of best fit')
plt.legend()

plt.show()
# Replace CES-D scores with change in volume
result_matrix=np.delete(result_matrix,-1,axis=1)

result_matrix=np.column_stack((result_matrix,change_in_volume))

# Initialize the scaler
scaler = StandardScaler()

model=LinearRegression()

# set training number and testing number
num_train = int(0.8*n)
num_test = int(0.2*n)
scaler.fit_transform(result_matrix)
# split data into training set and testing set
X_train = result_matrix[0:num_train,0:-1]
X_test = result_matrix[n-num_test:,0:-1]
Y_train = result_matrix[0:num_train,-1]
Y_test = result_matrix[n-num_test:,-1]

# Fit the model
model.fit(X_train, Y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the performance
mse = mean_squared_error(Y_test, y_pred)
r2 = r2_score(Y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
# Plot the predictions vs. actual values
plt.scatter(Y_test, y_pred)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Hippocampus Volume')


# Fit a line of best fit using numpy's polyfit
slope, intercept = np.polyfit(Y_test, y_pred, 1)

# Plot the line of best fit
plt.plot(Y_test, slope * Y_test + intercept, color='red', label='Line of best fit')
plt.legend()

plt.show()

from scipy.stats import ttest_ind
from scipy.stats import pearsonr

# Assuming data1 and data2 are your arrays
scaler = StandardScaler()
age_scaled = scaler.fit_transform(age_data.reshape(-1, 1))
ces_scaled = scaler.fit_transform(change_in_volume.reshape(-1, 1))


# Perform Pearson correlation
correlation_coefficient, p_value = pearsonr(change_in_volume, age_data)

# Print results
print("Effects age have on the change in Hippocampus Volume:")
print(f"Pearson Correlation Coefficient: {correlation_coefficient}")
print(f"P-value: {p_value}")

# Perform Pearson correlation
correlation_coefficient, p_value = pearsonr(change_in_volume, result_matrix[:,1])

# Print results
print("Effects sex have on the change in Hippocampus Volume:")
print(f"Pearson Correlation Coefficient: {correlation_coefficient}")
print(f"P-value: {p_value}")

# Perform Pearson correlation
correlation_coefficient, p_value = pearsonr(change_in_volume, weight_data)

# Print results
print("Effects weight have on the change in Hippocampus Volume:")
print(f"Pearson Correlation Coefficient: {correlation_coefficient}")
print(f"P-value: {p_value}")

# Perform Pearson correlation
correlation_coefficient, p_value = pearsonr(change_in_volume, height_data)

# Print results
print("Effects height have on the change in Hippocampus Volume:")
print(f"Pearson Correlation Coefficient: {correlation_coefficient}")
print(f"P-value: {p_value}")

# Function for calculating and classifying BMI
def categorize_bmi(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 24.9:
        return "Normal weight"
    elif 25.0 <= bmi < 29.9:
        return "Overweight"
    else:
        return "Obesity"
# Bmi array full of every participant
bmi_array = weight_data / (height_data ** 2)
# Print BMI and status for each person
for i, bmi in enumerate(bmi_array):
    status = categorize_bmi(bmi)
    print(f"Person {i + 1}:{status}")
```
